# Weekend Project: Scraping Ghibli Studio API

Main agenda of this project is to utilized the Ghibli Studio API (https://ghibliapi.herokuapp.com/) to scrape data and convert it to a csv file so you dont have to.

```python
import requests
import json
import pandas as pd
```

## Films

```python
films = requests.get(url = 'https://ghibliapi.herokuapp.com/films')
dic = json.loads(films.text)
```

```python
films_dataframe = pd.DataFrame.from_dict(dic)
films_dataframe.to_csv('films.csv')
films_dataframe.head()
```

## People

```python
people = requests.get(url = 'https://ghibliapi.herokuapp.com/people')
dic = json.loads(people.text)
```

```python
people_dataframe = pd.DataFrame.from_dict(dic)
people_dataframe.to_csv('people.csv')
people_dataframe.head()
```

## Locations

```python
locations = requests.get(url = 'https://ghibliapi.herokuapp.com/locations')
dic = json.loads(locations.text)
locations_dataframe = pd.DataFrame.from_dict(dic)
locations_dataframe.to_csv('locations.csv')
locations_dataframe.head()
```
